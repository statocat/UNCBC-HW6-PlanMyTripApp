```python
# make a list of coordinate pairs - the fast np way
lats = np.random.uniform(-90,90,1500)
lngs = np.random.uniform(-90,90,1500)
lats_lngs = zip(lats,lngs)
coords = list(lats_lngs)
coords
```

```python
# # print the coordinates
# for i in coords:
#     print(i[0],i[1])
```

```python
# use citipy to find the nearest cities to the random coordinates

for i in coords:
    print(citipy.nearest_city(i[0],i[1]).city_name,
    citipy.nearest_city(i[0],i[1]).country_code
    )
```

```python
# make a list of the nearest cities to coords elements and check to see at least 500

cities = []
for i in coords:
    city = citipy.nearest_city(i[0],i[1]).city_name
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
# Starting URL for Weather Map API Call.
from nturl2path import pathname2url


url = 'https://api.openweathermap.org/data/2.5/weather?units=Imperial&appid=' + ow_key
```

```python
city = 'Boston'
city_url = url + '&q=' + city
#city_url
```

```python
city_weather = requests.get(url + '&q=' + 'Boston')
pp.pprint(city_weather.json())
```

```python
if city_weather.status_code ==200:
    print(f'City weather found')
else:
    print(f'City weather not found')
```
